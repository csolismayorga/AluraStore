# Importación de datos



```python
import pandas as pdx

# Defina una lista de precios por categoría que se usará en los gráficos

categoría_precio_ordenado_lista=[]

# Invoca la función que crea lista de tiendas por cada url

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Se declara una lista de tiendas para optimizar la programación usando funciones

lista_tiendas = []
lista_urls = [url,url2,url3,url4]

# Esta función recibe una lista de urls para crear una lista de tiendas
# que será utilizada en los siguientes análisis

def Crea_Lista_Tiendas(lista_urls):
  for url in lista_urls:
    tienda = pdx.read_csv(url)
    lista_tiendas.append(tienda)

Crea_Lista_Tiendas(lista_urls)
```

#1. Análisis de facturación



```python
# El siguiente ciclo for barre la lista de tiendas para calcular el monto acumulado de ventas por cada tienda

total_facturado=0
print('--------------------------------------------------------')
for indice, tienda in enumerate(lista_tiendas):
  print(f'Monto total facturado de la tienda #{indice+1}: ',format(tienda.Precio.sum(),','))
  total_facturado += tienda.Precio.sum()

print('--------------------------------------------------------')
print('Monto total facturado general        : ',format(total_facturado,','))
print('--------------------------------------------------------')
```

# 2. Ventas por categoría

```python
# Función que imprime los totales de venta por categoría

def Imprime_Totales_x_Categoria(lista_tiendas):
    for indice, tienda in enumerate(lista_tiendas):
        print(f'Ventas por categoría de Tienda #{indice+1}')
        categoría_precio_ordenado=tienda.groupby('Categoría del Producto').sum()['Precio'].sort_values(ascending=False)
        print(categoría_precio_ordenado)
        categoría_precio_ordenado_lista.append(list(categoría_precio_ordenado))
        print(' ')

#Invoca la función Imprime_Totales_x_Categoria para imprimir totales por categoría de producto para cada tienda

Imprime_Totales_x_Categoria(lista_tiendas)
```

# 3. Calificación promedio de la tienda


```python
# Función que imprime la calificación promedio de cada tienda

def Imprime_CalificaciónPromedio_x_Tienda(lista_tiendas):
    for indice, tienda in enumerate(lista_tiendas):
        print(f"Calificación promedio de Tienda #{indice+1}: {format(tienda.Calificación.sum()/len(tienda),'.2f')}")

#Invoca la función Imprime_Totales_x_Categoria para imprimir totales por categoría de producto para cada tienda

Imprime_CalificaciónPromedio_x_Tienda(lista_tiendas)
```

# 4. Productos más y menos vendidos

```python
# Función que imprime los productos más vendidos de cada tienda

def Imprime_Productos_Más_Vendidos_x_Tienda(lista_tiendas):
    for indice, tienda in enumerate(lista_tiendas):
        print(f"\nLos 5 productos que menos dinero generaron de la tienda #{indice+1}:\n")
        producto_precio=tienda.groupby('Producto').sum()['Precio']
        producto_precio_ordenado=producto_precio.sort_values(ascending=False).head(5)
        print(producto_precio_ordenado)

# Función que imprime los productos menos vendidos de cada tienda

def Imprime_Productos_Menos_Vendidos_x_Tienda(lista_tiendas):
    for indice, tienda in enumerate(lista_tiendas):
        print(f"\nLos 5 productos que menos dinero generaron de la tienda #{indice+1}:\n")
        print(' ')
        producto_precio=tienda.groupby('Producto').sum()['Precio']
        producto_precio_ordenado=producto_precio.sort_values(ascending=False).tail(5)
        print(producto_precio_ordenado)

#Invoca la función Imprime_Totales_x_Categoria para imprimir totales por categoría de producto para cada tienda

print('----------------------------------------------------------- ')
print('       Los cinco productos más vendidos                     ')
print('----------------------------------------------------------- ')
print(' ')

Imprime_Productos_Más_Vendidos_x_Tienda(lista_tiendas)

#Invoca la función Imprime_Totales_x_Categoria para imprimir totales por categoría de producto para cada tienda

print(' ')
print('----------------------------------------------------------- ')
print('       Los cinco productos menos vendidos                   ')
print('----------------------------------------------------------- ')
print(' ')

Imprime_Productos_Menos_Vendidos_x_Tienda(lista_tiendas)
```

# 5. Envío promedio por tienda

```python
# Función que imprime el costo promedio de envío de cada tienda

def Imprime_CostoPromedioEnvío_x_Tienda(lista_tiendas):
    for indice, tienda in enumerate(lista_tiendas):
        Costo_promedio_envío_tienda = tienda['Costo de envío'].sum()/len(tienda)
        Costo_promedio_envío_tienda_str = "{:,.2f}".format(tienda['Costo de envío'].sum()/len(tienda))
        print(f"Costo de envío promedio de la tienda #{indice+1}: {Costo_promedio_envío_tienda_str}")

#Invoca la función Imprime_Totales_x_Categoria para imprimir totales por categoría de producto para cada tienda

Imprime_CostoPromedioEnvío_x_Tienda(lista_tiendas)
```

# 6. Gráficos

```python
#--------------------------------------------------------------------------------
# Se crea gráfico de columnas de ventas realizadas por cada tienda
#--------------------------------------------------------------------------------

import matplotlib.pyplot as plot

# Se crean dos listas

categorias=['Tienda #1','Tienda #2','Tienda #3','Tienda #4']
valores=[lista_tiendas[0].Precio.sum(),lista_tiendas[1].Precio.sum(),lista_tiendas[2].Precio.sum(),lista_tiendas[3].Precio.sum()]
colores = ['yellow', 'green', 'blue','red']

# Se genera gráfico de columnas

plot.bar(categorias,valores,color=colores)
plot.ylim(bottom=0) # Método funcional
plot.title('Monto Total Facturado por Tienda')
plot.xlabel('Tiendas')
plot.ylabel('Monto total facturado (USD)')
plot.show()

print('')

#--------------------------------------------------------------------------------
# Se crea gráfico comparativo de ventas por categoría de productos (Lineas)
#--------------------------------------------------------------------------------

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
categoría = ['Elect', 'ElectD', 'Muebles', 'Musica', 'Deporte', 'Juguetes', 'Hogar','Libros']
ventas = {'Tienda_1':categoría_precio_ordenado_lista[0],
          'Tienda_2':categoría_precio_ordenado_lista[1],
          'Tienda_3':categoría_precio_ordenado_lista[2],
          'Tienda_4':categoría_precio_ordenado_lista[3]
         }
ax.plot(categoría, ventas['Tienda_1'], color = 'tab:purple', label='Tienda 1')
ax.plot(categoría, ventas['Tienda_2'], color = 'tab:green',label='Tienda 2')
ax.plot(categoría, ventas['Tienda_3'], color = 'tab:orange',label='Tienda 3')
ax.plot(categoría, ventas['Tienda_4'], color = 'tab:red',label='Tienda 4')

# Añadir título y etiquetas a los ejes
plt.title('Gráfico de Ventas por Categoría de Producto')
plt.xlabel('Categoría de producto')
plt.ylabel('Monto facturado (USD)')

# Mostrar la leyenda
plt.legend()

plt.show()

print('')

#--------------------------------------------------------------------------------
# Se crea gráfico de barras calificación promedio por tienda
#--------------------------------------------------------------------------------

# Se crean dos listas

categorias=['Tienda #1','Tienda #2','Tienda #3','Tienda #4']
valores=[lista_tiendas[0].Calificación.sum()/len(lista_tiendas[0]),lista_tiendas[1].Calificación.sum()/len(lista_tiendas[1]),lista_tiendas[2].Calificación.sum()/len(lista_tiendas[2]),lista_tiendas[3].Calificación.sum()/len(lista_tiendas[3])]

# Se genera gráfico

plot.barh(categorias,valores,color=colores)
plot.title('Calificación Promedio por Tienda')
plot.xlabel('Tiendas')
plot.ylabel('Calificación Promedio')
plot.show()

print('')

#--------------------------------------------------------------------------------
# Se crea gráfico de barras costo de envío promedio de productos por tienda
#--------------------------------------------------------------------------------

# Se crean dos listas

categorias=['Tienda #1','Tienda #2','Tienda #3','Tienda #4']
valores=[lista_tiendas[0]['Costo de envío'].sum()/len(lista_tiendas[0]),lista_tiendas[1]['Costo de envío'].sum()/len(lista_tiendas[1]),lista_tiendas[2]['Costo de envío'].sum()/len(lista_tiendas[2]),lista_tiendas[3]['Costo de envío'].sum()/len(lista_tiendas[3])]

# Se genera gráfico

plot.barh(categorias,valores,color=colores)
plot.title('Costo Promedio de Envío por Tienda')
plot.xlabel('Tiendas')
plot.ylabel('Costo Promedio de Envío (USD)')
plot.show()

print('')

#--------------------------------------------------------------------------------
# Se crea gráfico de pie de costo de envío promedio de productos por tienda
#--------------------------------------------------------------------------------

# Datos para el gráfico (valores de los segmentos)

sizes =[(lista_tiendas[0]['Costo de envío'].sum()/len(lista_tiendas[0])),(lista_tiendas[1]['Costo de envío'].sum()/len(lista_tiendas[1])),(lista_tiendas[2]['Costo de envío'].sum()/len(lista_tiendas[2])),(lista_tiendas[3]['Costo de envío'].sum()/len(lista_tiendas[3]))]

# Etiquetas para los segmentos
labels = ['Tienda #1', 'Tienda #2', 'Tienda #3', 'Tienda #4']

# Crea el gráfico de pie
plot.pie(sizes, labels=labels, autopct='%1.1f%%')  # autopct muestra los porcentajes

# Añade un título
plot.title('Gráfico de Costos Promedio de Envío por Tienda')

# Muestra el gráfico
plot.show()

```
